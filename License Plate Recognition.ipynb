```python
def load_image_and_resize_with_bb(image_path, xmin, xmax, ymin, ymax):

    # Read the Images Using CV2
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Read the Height, Width, and Channel
    h, w, _ = img.shape

    # Resize the image
    img_resized = cv2.resize(img, (255, 255))

    # Scale the bounding box values to the new image size
    xmin_scaled = int(xmin * 255/w)
    xmax_scaled = int(xmax * 255/w)
    ymin_scaled = int(ymin * 255/h)
    ymax_scaled = int(ymax * 255/h)

    # Scaled Bounding Box Coordinates
    coordinates = (xmin_scaled, xmax_scaled, ymin_scaled, ymax_scaled)

    return img_resized, coordinates
```

```python
image = []
bb_coordinates = []

for image_name in os.listdir(vehicle_train_dataset):

    # Make a Image Path
    image_path = os.path.join(vehicle_train_dataset, image_name)

    # Bounding Box Coordinates of the Image
    image_bb = bounding_box_df[bounding_box_df['img_id'] == image_name]

    # Extract the Bounding Box Coordinates
    xmin, xmax, ymin, ymax = int(image_bb['xmin']), int(image_bb['xmax']), int(image_bb['ymin']), int(image_bb['ymax'])

    # Iterate each Image to get Image_array and Bounding_Box_Coordinates
    img, coordinates = load_image_and_resize_with_bb(image_path, xmin, xmax, ymin, ymax)

    image.append(img)
    bb_coordinates.append(coordinates)
```

```python
# Define a Function process to Split Dataset into Training(80%), Validation(10%) and Testing(10%) Sets

def train_validation_test_split(dataset, train_size=0.8, validation_size=0.1, test_size=0.1, shuffle=True, shuffle_size=10000):

    dataset_batch_count = len(dataset)                                                  # Batch Images dataset Length is 57
    train_batch_count = int(dataset_batch_count * train_size)                           # int(57*0.8) ---> 45 Batches are Training
    validation_test_batch_count = int(dataset_batch_count * validation_size)            # int(57*0.1) ---> 5 Batches are validation & Remaining Testing

    if shuffle:
        dataset = dataset.shuffle(buffer_size=shuffle_size)                             # Suffling the Dataset help to Model Understand the Data Well

    train_ds = dataset.take(train_batch_count)                                          # First 45 Batches are Training_Dataset
    validation_ds = dataset.skip(train_batch_count).take(validation_test_batch_count)   # Remaining 12 Batches ---> Next 5 Batch are Validation_Dataset
    test_ds = dataset.skip(train_batch_count).skip(validation_test_batch_count)         # Pending 7 Batches are Testing

    return train_ds, validation_ds, test_ds
```

```python
# Apply the Function in Potato Dataset
train_ds, validation_ds, test_ds = train_validation_test_split(dataset)
len(train_ds), len(validation_ds), len(test_ds)
```

### **Model Building and Training**

```python
# Build a Data Augmentation Pipeline

augmentation_pipeline = keras.Sequential([
                                keras.layers.Resizing(height=255, width=255),
                                keras.layers.Rescaling(scale=1./255),
                                keras.layers.RandomFlip(mode='horizontal_and_vertical'),
                                keras.layers.RandomRotation(factor=0.2)
                            ])
augmentation_pipeline
```

```python
# Define the Parameters of Input_shape, Channel and Target

channel = 3       # RGB ---> Color Images
target  = 4       # Bounding Box Coordinates
image_size = 255  # Unique Size of All Images

input_shape = (None, image_size, image_size, channel)   # Automatically calculate the Batch Size of Dataset
input_shape
```

```python
# Build a CNN Model Architecture

model = keras.Sequential([

            # Preprocessing Layer
            augmentation_pipeline,

            # Input Layer
            keras.layers.Input(shape=input_shape),

            # 1st Conv Layer
            keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'),
            keras.layers.MaxPooling2D(pool_size=(2, 2), padding='same'),

            # 2nd Conv Layer
            keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
            keras.layers.MaxPooling2D(pool_size=(2, 2), padding='same'),

            # 3rd Conv Layer
            keras.layers.Conv2D(filters=128, kernel_size=(3,3), activation='relu'),
            keras.layers.MaxPooling2D(pool_size=(2, 2), padding='same'),

            # Flatten to 1-Dimentional Vector
            keras.layers.Flatten(),

            # 1st Fully Connected Dense Layer
            keras.layers.Dense(units=64, activation='relu', kernel_regularizer=keras.regularizers.L2(l2=1e-2)),
            # keras.layers.Dropout(0.3),

            # 2nd Fully Connected Dense Layer
            keras.layers.Dense(units=128, activation='relu', kernel_regularizer=keras.regularizers.L2(l2=1e-2)),
            # keras.layers.Dropout(0.3),

            # Output Layer
            keras.layers.Dense(units=target)
        ])
model
```

```python
# Display the Model Summary

model.build(input_shape)
model.summary()
```

```python
# Compiling the Model with Optimizer, Loss, and Metrics

model.compile(optimizer='adam',
              loss='mse',
              metrics=['mae', keras.metrics.RootMeanSquaredError()])
```

```python
# Model Training with Training_Dataset and Evaluated with Validation_Dataset

history = model.fit(train_ds,
                    batch_size=batch_size,
                    epochs=200,
                    verbose=1,
                    validation_data = validation_ds)
history
```

```python
# Evaluating Model Performance on Test Dataset
model.evaluate(test_ds)
```
